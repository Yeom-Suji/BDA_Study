```python
import seaborn as sns
planets= sns.load_dataset('planets')
planets.shape
```

```python
planets.head()
```

```python
decade= 10 * (planets['year'] // 10)
decade= decade.astype(str) + 's'
decade.name= 'decade'
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0) #'method' 및 'decade'에 대한 그룹화를 수행하고, 'number' 열의 합계를 계산한 후 결과를 재구성하고 누락된 값을 0으로 채움.
```
