```python
import numpy as np

mean= [0,0] #다변수 정규 분포의 평균 벡터
cov= [[1,2], #다변수 정규 분포의 공분산 행렬
     [2,5]]
X= np.random.multivariate_normal(mean, cov, 100)
X.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() #플롯 스타일 설정
plt.scatter(X[:,0], X[:,1]);
```

```python
#팬시인덱싱으로 임의의 점 20개 선택
indices= np.random.choice(X.shape[0], 20, replace=False)
indices
```

```python
selection= X[indices]
selection.shape
```

```python
#선택한 점 동그라미 표시
plt.scatter(X[:,0],X[:,1],alpha=0.3)
plt.scatter(selection[:,0],selection[:,1], facecolor='none', edgecolor='red', s=200);
```
